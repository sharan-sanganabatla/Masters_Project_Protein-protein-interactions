# The CSV file containing raw data is given as input and this python file  does the basic preprocessing, splits the dataset into training and test and generates .pkl fikles for prediction algorithm.

```python
import pandas as pd
df = pd.read_csv("begin.csv")
df
```

```python
df=df.rename(columns={'yersinia_sequence':'pathogen_sequence'})
```

```python
df["human.pid"] = df["human.pid"].map(lambda x: x[10:])
df["influenza.pid"] = df["influenza.pid"].map(lambda x: x[10:])
```

```python
df
```

```python
df[df.duplicated()]
```

```python
df = df.drop_duplicates(keep=False)
df.drop(df.tail(1).index,inplace=True) 
df
```

```python
df['length']= df['human_sequence'].apply(len)
df
```

```python
df.drop(df[df.length < 35].index, inplace=True)
df
```

```python
df_join = df["human_sequence"]+df["pathogen_sequence"]
df_join
```

```python
df_human = df["human_sequence"]
df_human
```

```python
df_influenza = df["pathogen_sequence"]
df_influenza
```

```python
list1 = []
list2 = []
list3 = []
for i,j in df_join.items():
    out = []
    out = [(j[i:i+1]) for i in range(0,len(j),1)]
    list1.append(out)
for i,j in df_human.items():
    out = []
    out = [(j[i:i+1]) for i in range(0,len(j),1)]
    list2.append(out)
for i,j in df_influenza.items():
    out = []
    out = [(j[i:i+1]) for i in range(0,len(j),1)]
    list3.append(out)
```

```python
series1 = pd.Series(list1)
series2 = pd.Series(list2)
series3 = pd.Series(list3)
```

```python
df3 = pd.DataFrame({'Human':series2,'Pathogen':series3,'Joined':series1})
df3
```

```python
df3 = df3.sample(frac=1)
df3
```

```python
df3 = df3.reset_index()
df3 = df3.drop('index', 1)
df3
```

```python
count1 = df3.shape[0]
if count1 % 2 == 0:
    one=zero=int(count1/2)
else:
    one = int(count1/2)+1
    zero= one-1
labelList = [1]*one+[0]*zero
```

```python
import random
random.shuffle(labelList)
df3['label'] = labelList
df3
```

```python
import pickle
```

```python
df3.to_pickle("formatted.pkl")
```
